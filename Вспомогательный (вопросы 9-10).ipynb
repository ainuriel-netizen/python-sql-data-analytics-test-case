```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
listers = pd.read_csv('listers.csv')
revenue_grouped = listers.groupby('user_id')['revenue'].sum()
revenue_grouped
```

```python
sns.histplot(data=revenue_grouped)
plt.show()
```

```python
age_grouped = listers.groupby('user_id')['age'].min()
age_grouped.median()
```
